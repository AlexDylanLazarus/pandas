# How to create a column in dataframe

```python
tips_df['tips_per_person']=tips_df['tip']/tips_df['size']
tips_df.head(5) #get first top 5
tips_df.tail(10) #gives the last 10
```

find how much percentage peple are tiping
- percent 
- add it as another column 


```python
tips_df['tip_percentage'] = round(((tips_df['tip']/tips_df['total_bill'])*100),2)
tips_df
```

```python
tips_df.sort_values(by="total_bill", ascending=False)
tips_df.sort_values(by="tip_percentage", ascending=False).head(5)
```

# task 
Find the rows with the highest tip_percent skip the first 5 on a saturday

```python
filter_me = tips_df[(tips_df['day'] == "Sat") | (tips_df['day'] == 'Sun')].sort_values(by="tip_percentage",ascending=False).iloc[5::]
filter_me

#or
#or
tips_df[tips_df['day'].isin(['Sun','Sat'])].sort_values(by="tip_percentage",ascending=False).iloc[5::].loc[:,['tip_percentage','day','tip','total_bill']]
```
